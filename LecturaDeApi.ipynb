```python
#Siempre vemos primero la api que querramos con un JSON VIEWER
#http://jsonviewer.stack.hu/
from urllib.request import urlopen
import json 
import pandas as pd 

```

```python
#Traemos entonces primero la url con la api que queremos trabajar 
url="https://jsonplaceholder.typicode.com/users"
#Luego abro esa url con la variable correspondiente 

response=urlopen(url)
```

```python
#Asi que queremos leer con la apertura hecha el objeto en json  
contenido = json.loads(response.read())
```

```python
#Ya tenemos entonces todo el contenido del json
print(contenido)
```
